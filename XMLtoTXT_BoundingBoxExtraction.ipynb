```python
import csv 
import xml.etree.ElementTree as ET 
from os import listdir
import os

import numpy as np
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
from PIL import Image
from keras.models import load_model

import pandas as pd
```

```python
parseFiles = [f for f in listdir('TrainData/annotations/') if f.split('.')[1] == 'xml']

print(len(parseFiles))
```

```python
file = 'TrainData/annotations/'+ parseFiles[0]
tree = ET.parse(file)
root = tree.getroot()
```

```python
labelMeta = ['filepath', 'xmin', 'ymin', 'xmax', 'ymax', 'class_name']
imageMeta = ['filepath', 'width', 'height', 'depth']

labels = pd.DataFrame(columns=labelMeta)
image_data = pd.DataFrame(columns=imageMeta)

appends = ['xmin', 'ymin', 'xmax', 'ymax', 'class_name']

countClass = 0
countFiles = 0
ripe = 0
unripe = 0

for strawberryXML in parseFiles:
    file = 'TrainData/annotations/'+ strawberryXML
    
    tree = ET.parse(file)
    root = tree.getroot()
    countFiles +=1
    fileName = root.find('filename').text
    
    imagePath = 'TrainData/images/'+fileName
    
    imageWidth = root.find('size/width').text
    imageHeight = root.find('size/height').text
    imageDepth = root.find('size/depth').text
    
    temp = [imagePath, imageWidth, imageHeight, imageDepth]
    
    image_data=image_data.append(pd.DataFrame([temp], columns=imageMeta))
    
    for i in root.iter('object'):
        binaryClass = i[0].text
        #print(binaryClass)
        countClass +=1
        
        if binaryClass == 'ripe_strawberry':
            ripe += 1
        else:
            unripe +=1
        temp2 = [imagePath]
        for n,j in enumerate(list(i[5])):
            temp2.append(int(j.text))
#             if imagePath == "TrainData/annotations/strawimage175.jpg":
#                 print(int(j.text))
        temp2.append(binaryClass)
        labels=labels.append(pd.DataFrame([temp2], columns=labelMeta))
            
print('Total number of images processed =', countFiles)
print('Total number of unripe strawberries =', unripe)
print('Total number of ripe strawberries =', ripe)
assert(ripe + unripe == countClass)
```
