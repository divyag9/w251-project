```python
images = [f for f in os.listdir('TestAcrossEpochs') if f[-4:] == '.png']
```

```python
for file in images:
    names = file.split('-')
    epoch = re.sub("[^0-9]", "", names[-1])
    os.rename('TestAcrossEpochs/{}'.format(file),'TestAcrossEpochs/{}-{}.png'.format(names[0],epoch))
```

```python
images = [f for f in os.listdir('TestAcrossEpochs') if f[-4:] == '.png']
images.sort(key = lambda x: int(x.split('-')[1].split('.')[0]))
images
```

```python
unique_images = []

for image in images:
    if image.split('-')[0] not in unique_images:
        unique_images.append(image.split('-')[0])
```

```python
unique_images
```

```python
sorted_images = [[] for _ in unique_images]

for image in images:
    name = image.split('-')[0]
    indice = unique_images.index(name)
    sorted_images[indice].append(imageio.imread('TestAcrossEpochs/{}'.format(image)))
```

```python
for j,i in enumerate(unique_images):
    imageio.mimsave('TestImageGifs/{}.gif'.format(i), sorted_images[j],duration=0.35)
```
